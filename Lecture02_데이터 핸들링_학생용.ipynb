넘파이(NumPy)
-------------
머신러닝의 주요 알고리즘은 선형대수와 통계 등에 기반한다. 특히 선형대수는 수학뿐만 아니라 다른 영역의 자연과학, 공학에서 널리 쓰인다. 이러한 선형대수를 프로그램에서 쉽게 다룰 수 있도록 지원하는 대표적인 패키지가 넘파이(NumPy)이다.

넘파이(NumPy)의 기반의 데이터 타입은 ndarray이다.
- array(리스트): 리스트와 같은 인자를 받아 ndarray로 변환한다.
- shape: ndarray의 크기, 행과 열의 수를 튜플 형태로 가지는 객체 변수 (차원또한 알 수 있다. )
- ndim : ndarray의 차원 정보를 가지는 객체 변수

ndarray내의 데이터 값은 숫자, 문자, 불(참/거짓) 값 모두 가능하다.<br/>
ndarray의 계산의 특성상 동일한 ndarray내의 값은 모두 데이터 타입을 가진다.<br/>
만약 다른 데이터 타입을 가지는 리스트일 경우 크기가 더 큰 데이터 타입으로 변경된다.
- dtype: ndarray내 값의 데이터 타입을 가지는 객체 변수

메모리를 절약하기 위해 narray내 값의 데이터 타입을 강제로 변경해줘야 할 필요가 있다.
- astype('데이터 타입'): 파라미터로 변경할 데이터 타입을 문자형으로 입력해준다. ndarray내 값을 해당 데이터 타입으로 변경한다.

자주 사용되는 형태의 ndarray은 특정 함수를 통해 쉽게 선언할 수 있다.
- arange(정수): 0부터 (입력받은 정수-1)까지 순차적으로 구성된 ndarray를 반환한다.
- arange(start=정수, stop=정수): start부터 (stop-1)까지 순차적으로 구성된 ndarray를 반환한다.
- zeros(정수): 0부터 (입력받은 정수-1)까지 0으로 구성된 ndarray를 반환한다.
- ones(정수): 0부터 (입력받은 정수-1)까지 1로 구성된 ndarray를 반환한다.
 <br/>
**arange()로 생성된 ndarray내 값의 데이터 타입은 int64** <br/>
**zeros(), ones()로 생성된 ndarray내 값의 데이터 타임은 float64**

ndarray의 차원과 크기를 변경해야할 필요가 있다. reshape() 메소드를 통해 해당 작업을 수행 할 수 있다.
- reshape(리스트): ndarray를 입력받은 shape으로 변경한다. -1을 파라미터로 입력하면 다른 파라미터와 호환되는 shape으로 변경한다.

**만약 입력받은 shape으로 변경하지 못할 경우 에러가 발생한다.**

ndarray내 값에 접근하기 위해서는 리스트와 유사하게 대괄호([]) 안에 인덱스를 넣으면 된다.<br/>
list와 마찬가지로 특정 값에 접근하여 변경할 수 있다.

ndarray내 특정 값을 세트롤 추출하는 방법은 다음과 같다.
- 특정 인덱스를 통한 추출: 대괄호([]) 안에 인덱스를 넣어 값을 추출한다.
- 슬라이싱(slicing): 연속된 인덱스상의 값을 ndarray로 추출한다. ':' 앞뒤에 시작 인덱스 종료 인덱스를 입력한다. 만약 적지 않으면 해당 방향으로 끝까지 ndarray로 추출한다.
- 팬시 인덱싱(Fancy Indexing): 일정한 인덱싱 집합을 리스트로 지정해 ndarray로 추출한다.
- 불린 인덱싱(Boolean Indexing): 조건식을 만족하는 값을 ndarray로 추출한다.

행렬을 정렬하는 방법은 두가지가 존재한다.
- numpy.sort(<ndarray>): 함수로 파라미터로 입력된 ndarray가 정렬된 값을 반환한다. **원본은 그대로 유지된다.**
- ndarray.sort(): 메소드로 원본을 정렬한다. 반환값이 없다(None). **원본이 유지되지 않는다.**

정렬된 행렬의 값들의 원본 인덱스가 필요한 경우가 있다. arsort()함수를 통해 해당 작업을 수행할 수 있다.
- argsort(): 행렬 정렬 시 원본 행렬의 인덱스를 ndarray로 반환한다.
