머신러닝
========
머신러닝은 문제 해결을 위한 알고리즘으로 사용된다. 일반적인 지도 학습의 경우 문제와 정답을 동시에 주고 규칙성을 스스로 파악하도록 한다. 사이킷런(Scikit-Learn)은 여러 가지 종류의 머신러닝 알고리즘들을 통해 사용자들이 머신러닝을 쉽게 사용할 수 있도록 도와준다. 특히 분류, 회귀, 군집화, 모델 선택 및 최적화 등의 작업과 메트릭 학습과 유사한 최적화 기법을 사용하는데 유용하다. 이러한 방법들로 머신러닝 작업을 더 효율적으로 수행할 수 있게 도와준다.

머신러닝은 다음과 같은 순서로 진행된다.
- 데이터 수집 및 불러오기
- 데이터 전처리
- 데이터 분류 (학습용/테스트용)
- 학습하기
- 예측하기
- 예측 정확도 분석

**붓꽃 품종 예측하기**를 통해 머신러닝의 대략적 과정을 이해해 보자

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

사이킷런에 내장된 데이터 세트 구성
- data: 특징 데이터 세트
- target: 라벨 데이터 , 숫자형 데이터

- target_names: 개별 라벨의 이름
- DESCR: 데이터 세트에 대한 설명과 각 피쳐에 대한 설명

```python
import pandas as pd

# 붓꽃 데이터 세트 중 특징 데이터
iris_data = iris.data
iris_df = pd.DataFrame(iris_data, columns=iris.feature_names)
iris_df.head()
```

```python
# 학습용 데이터 테스트용 데이터 분리
''' X_train: 학습용 특징 데이터
    X_test: 테스트용 특징 데이터
    y_train: 학습용 라벨 데이터
    y_test: 테스트용 라벨 데이터 '''

# 난수 생성
import random
from time import time
random.seed(time())

random_value = int(random.random() * 10000)

# 랜덤하게 80%는 학습용 데이터, 20%는 테스트 데이터로 분리
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, random_state = 25)
```

```python
# 학습 수행
dt_clf.fit(X_train, y_train)
```

```python
# 예측 수행
pred = dt_clf.predict(X_test)
```

```python
# 예측 정확도 분석
from sklearn.metrics import accuracy_score

print('실제 값: ', y_test)
print('예측 값: ', pred)
print("예측 정확도: {0:.4f}".format(accuracy_score(y_test, pred)))
```

----------------------------

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 붓꽃 데이터 불러오기
iris = load_iris()
dt_clf = DecisionTreeClassifier()
data = iris.data
label = iris.target

# 학습하기
dt_clf.fit(data, label)

# 예측하기
pred = dt_clf.predict(data)
print("예측 정확도: {0:.4f}".format(accuracy_score(label, pred)))
```

위의 예측 결과가 1 즉, 100%인 이유는 이미 학습한 데이트 세트를 기반으로 예측을 진행했기 때문이다. 주어진 자료에만 과도하게 최적화되었기에 새로운 데이터가 주어지면 낮은 정확도를 보여주기도한다. 이러한 문제를 과적합 문제라고 한다. 이를 방지하고자 학습/테스트 데이터를 분리한다.

train_test_split(특징 데이터 세트, 라벨 데이터 세트, test_size=0.25, shuffle=True, random_state):<br/>
test_size: 테스트 데이터의 비율을 입력한다. (0 ~ 1) <br/>
shuffle: 주어진 데이터를 섞을 것인지 결정한다. <br/>
random_state: 데이터를 섞을 때 사용할 시드값

교차검증
--------
고정된 학습 데이터와 테스트 데이터로 평가하다보면 고정된 데이터에 과적합되는 문제가 발생한다. 이를 예방하기 위해 다양하게 학습 데이터와 테스트 데이터를 나누는 교차검증을 진행한다.

**K 폴드 교차 검증** <br/>
가장 보편적으로 사용되는 교차 검증 기법으로 먼저 K개의 데이터 폴드 세트를 만들어 K번 만큼 각 폴드 세트에 학습과 검증을 반복적으로 진행한다.

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

# 붓꽃 데이터 불러오기
iris = load_iris()
dt_clf = DecisionTreeClassifier(random_state = 2023)
features = iris.data
label = iris.target

# 5개의 폴드 세트로 분리하기 위해 KFold 객체 생성
kfold = KFold(n_splits=5)

# 각 폴드 별 정확도
accuracy = []
```

```python
n_iter = 0

# KFold 객체의 split() 메소드를 호출하면 폴드 별 학습용, 검증요 테슽의 로우 인덱스 반환
for train_index, test_index in kfold.split(features):
    # KFold.split()으로 얻은 인덱스로 데이터 추출
    X_train = features[train_index]
    X_test = features[test_index]
    y_train = label[train_index]
    y_test = label[test_index]

    # 학습
    dt_clf.fit(X_train, y_train)

    # 예측
    pred = dt_clf.predict(X_test)

    # 정확도
    accuracy.append(accuracy_score(y_test, pred))
    train_size = X_train.shape[0] # 행의 수 == 데이터의 수
    test_size = X_test.shape[0] # 행의 수 == 데이터의 수
    print('{0}번째 교차 검증 정확도: {1:.4f}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'
    .format(n_iter+1, accuracy[n_iter], train_size, test_size))

    n_iter += 1
    print('{0}번째 교차 검증 데이터 인덱스: '.format(n_iter), test_index)
    print('학습 데이터 분포: \n', pd.DataFrame(y_train).value_counts(), '\n')

print('평균 검증 정확도: ', np.mean(accuracy))
```

붓꽃 품종는 각각 50개씩 동일한 수를 가진다. 하지만 위의 방식대로 폴드를 구분하게 되면 각 폴드에 들어있는 품종데이터는 전체 데이터의 비율과 달라지는 문제가 발생하게 된다. 이를 방지하고자 Stratified K 폴드를 이용한다.

**Stratified K 폴드**<br/>
불균형한 분포도를 가진 데이터 집합을 위한 K 폴드 방식이다. Stratified K 폴드는 전체 데이터 세트에서 각 데이터의 비율과 폴드 내 데이트의 비율이 동일하도록 데이터 세트를 분배한다.

```python
from sklearn.model_selection import StratifiedKFold

# 5개의 균일한 폴드 세트로 분리하기 위해 StratifiedKFold 객체 생성
skf = StratifiedKFold(n_splits = 5)

# 붓꽃 데이터 불러오기
iris = load_iris()
dt_clf = DecisionTreeClassifier(random_state = 600)
features = iris.data
label = iris.target
n_iter = 0

# 각 폴드 별 정확도
accuracy = []

# label 내 데이터 비율에 따라 나누기에 label 또한 입력
for train_index, test_index in skf.split(features, label):
    # StratifiedKFold.split()으로 얻은 인덱스로 데이터 추출
    X_train = features[train_index]
    X_test = features[test_index]
    y_train = label[train_index]
    y_test = label[test_index]

    # 학습
    dt_clf.fit(X_train, y_train)

    # 예측
    pred = dt_clf.predict(X_test)

    # 정확도
    accuracy.append(accuracy_score(y_test, pred))
    train_size = X_train.shape[0] # 행의 수 == 데이터의 수
    test_size = X_test.shape[0] # 행의 수 == 데이터의 수
    print('{0}번째 교차 검증 정확도: {1:.4f}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'
    .format(n_iter+1, accuracy[n_iter], train_size, test_size))

    n_iter += 1
    print('{0}번째 교차 검증 데이터 인덱스: '.format(n_iter), test_index)
    print('학습 데이터 분포: \n', pd.DataFrame(y_train).value_counts(), '\n')

print('평균 검증 정확도: ', np.mean(accuracy))
```

데이터 전처리
-------------
머신러닝 알고리즘에서 NaN값과 문자열 값을 허용하지 않는다. 결손값은 적절한 값으로 채워넣고 문자열 값은 모두 인코딩하여 숫자형 데이터로 변환해야한다. 이러한 과정을 데이터 전처리라 한다.

**레이블 인코딩(Lable encoding)**<br/>
카테고리 특징을 코드형 숫자 값으로 변환하는 것이다. 예를 들어 과일이 '포도', '감', '사과', '수박', '배'가 있다면 이들을 각각 '포도':0, '감':1, '사과':2, '수박':3, '배':4와 같은 숫자형으로 변환한다.

```python
from sklearn.preprocessing import LabelEncoder

fruits = ['포도', '감', '사과', '사과', '수박', '배', '포도', '배']

# 레이블 인코딩
encoder = LabelEncoder()
encoder.fit(fruits)
label = encoder.transform(fruits)
print("레이블 인코딩: ", label)
```

```python
# classes_ 객체 값을 통해 어떤 문자열이 어떤 숫자로 변환 됐는 지 확인할 수 있다.
encoder.classes_
```

```python
# inverse_transform() 메소드를 통해 원본 값으로 복구할 수 있다.
print('원본: ', encoder.inverse_transform(label))
```

1보다 2는 더욱 큰 수 이기에 일부 머신러닝 알고리즘에서 '배'보다 '사과'에 더욱 큰 가중치를 두는 문제가 발생하기도 한다. 이러한 문제점은 원-핫 인코딩을 통해 해결할 수 있다.

**원-핫 인코딩(One-Hot Encoding)**<br/>
각 특징에 따른 컬럼을 만들고 해당 특징에 해당하면 1 그렇지 않으면 0을 표시하는 방식이다.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import numpy as np

fruits = ['포도', '감', '사과', '사과', '수박', '배', '포도', '배']

# 먼저 숫자 값으로 변환하기 위해 레이블 인코딩을 적용한다.
encoder = LabelEncoder()
encoder.fit(fruits)
label = encoder.transform(fruits)

# 2차원 데이터로 변환
label = label.reshape(-1, 1)

# 원-핫 인코딩 적용
onehot = OneHotEncoder()
onehot.fit(label)
onehot_label = onehot.transform(label)

print('원-핫 인코딩 결과')
onehot_label.toarray()
```

서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업을 피처 스케이링(Feature Scaling)이라한다. 대표적으로 표준화(Standardization), 정규화(Normalization)가 있다.
- 표준화(Standardizatioin): 특징 데이터 각각을 평균이 0이고 분산이 1인 가우시안 정규 분포를 가진 값으로 변환하는 것
- 정규화(Normalization): 서로 다른 특징 데이터의 크기를 통일하기 위해 크기를 변환하는 것

```python
# 붓꽃 데이터 불러오기
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(iris_data, columns=iris.feature_names)
```

**StandardScaler**<br/>
표준화를 지원하는 모듈, 평균이 0, 분산이 1인 값으로 변환한다.

```python
# 평균, 분산 값 확인
print('평균값:')
print(iris_df.mean(), '\n') # mean(): 평균값 계산 메소드

print('분산값: ')
print(iris_df.var()) # var(): 분산값 계산 메소드
```

```python
# 표준화
from sklearn.preprocessing import StandardScaler

# StandardScaler 객체 생성
st_scaler = StandardScaler()

# StandatdScaler로 데이터 세트 변환
st_scaler.fit(iris_data) # ndarray 또는 DataFrame을 입력할 수 있다.
iris_scaled = st_scaler.transform(iris_data) # ndarray 또는 DataFrame을 입력할 수 있다. 반환값은 ndarray이다.

# 반환된 ndarray값을 DataFrame으로 변환
scaled_df =  pd.DataFrame(iris_scaled, columns=iris.feature_names)

# 평균, 분산 값 확인
print('평균값:')
print(scaled_df.mean(), '\n') # mean(): 평균값 계산 메소드

print('분산값: ')
print(scaled_df.var()) # var(): 분산값 계산 메소드
```

**MinMaxScaler**<br/>
정규화를 지원하는 모듈, 0\~1의 범위 값으로 변환한다. (음수가 있으면 -1\~1의 범위 값으로 변환한다.)

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler 객체 생성
mm_scaler = MinMaxScaler()

# MinMaxScaler로 데이터 세트 변환
mm_scaler.fit(iris_data)  # ndarray 또는 DataFrame을 입력할 수 있다.
iris_scaled = mm_scaler.transform(iris_data) # ndarray 또는 DataFrame을 입력할 수 있다. 반환값은 ndarray이다.

# 반환된 ndarray값을 DataFrame으로 변환
scaled_df =  pd.DataFrame(iris_scaled, columns=iris.feature_names)

# 최대, 최소 값 확인
print('최솟값:')
print(scaled_df.min(), '\n') # min(): 최솟값 추출 메소드

print('최댓값: ')
print(scaled_df.max()) # max()):최댓값 추출 메소드
```

'Age' 컬럼의 경우 값의 종류가 많기 때문에 범위별 카테고리로 분류한다.

```python
# 나이에 따라 구분 값을 반환하는 함수
''' Baby: 0~5세
    Child: 6~12세
    Teenager: 13~18세
    Student: 19~25세
    Yong Adult: 26~35세
    Adult:36~59세
    Elderly: 60세 이상 '''

def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Sutdent'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 59: cat = 'Adult'
    else : cat = 'Ederly'

    return cat

pp_df['Age'] = pp_df['Age'].apply(lambda x : get_category(x))
```

**레이블인코딩**<br/>
'Sex', 'Cabin', 'Embarked' 컬럼은 문자열 데이터이기에 레이블 인코딩을 이용해 숫자형 데이터로 변환한다.

```python
# 라벨 인코딩 진행
from sklearn.preprocessing import LabelEncoder

labels = ['Sex', 'Embarked']

for feature in labels:
    encoder = LabelEncoder()
    encoder.fit(pp_df[feature])
    pp_df[feature] = encoder.transform(pp_df[feature])
```

```python
encoder = LabelEncoder()
encoder.fit(pp_df['Cabin'])
pp_df['Cabin'] = encoder.transform(pp_df['Cabin'])
```

**원-핫 인코딩**<br/>
'Age' 컬럼의 경우 데이터 Baby, Child, Teenager, Student, Young Adult, Adult , Ederly로 구분할 수 있다. 또한 일부 머신러닝 알고즘에서 특정 카테고리에 높은 가중치를 둘 수 있기에 원-핫 인코딩을 적용한다.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

# 라벨 인코딩 진행
lb_encoder = LabelEncoder()
lb_encoder.fit(pp_df['Age'])
pp_df['Age'] = lb_encoder.transform(pp_df['Age'])

# 원-핫 인코딩 진행
# 나이 카데고리 데이터 추출
age_cat = pp_df['Age'].values
# 2차원 데이터로 변환
age_cat = age_cat.reshape(-1, 1)

# 원-핫 인코딩 진행
onehot = OneHotEncoder()
onehot.fit(age_cat)
age_data = onehot.transform(age_cat).toarray()


age_df = pd.DataFrame(age_data, columns=lb_encoder.classes_)
age_df
```

```python
pp_df = pd.concat([pp_df, age_df], axis=1)
pp_df.drop('Age', axis=1, inplace=True)
pp_df
```

**정규화, MinMaxScaler**<br/>
'Fare'컬럼만 숫자 데이터의 값이 크다. 이는 'Fare' 컬럼에 높은 가중치를 줄 수 있기에 정규화를 진행한다.

###데이터 분류 (학습용/테스트용)

```python
from sklearn.model_selection import StratifiedKFold

titanic_label = pp_df['Survived']
titanic_data = pp_df.drop('Survived', axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dt_clf = DecisionTreeClassifier(random_state = 213)
```

```python
# 5개의 균일한 폴드 세트로 분리하기 위해 StratifiedKFold 객체 생성
skf = StratifiedKFold(n_splits = 5)

# 각 폴드 별 정확도
accuracy = []



# label 내 데이터 비율에 따라 나누기에 label 또한 입력
for train_index, test_index in skf.split(titanic_data, titanic_label):
    # StratifiedKFold.split()으로 얻은 인덱스로 데이터 추출
    X_train = titanic_data.iloc[train_index]
    X_test = titanic_data.iloc[test_index]
    y_train = titanic_label.iloc[train_index]
    y_test = titanic_label.iloc[test_index]

    # 학습
    dt_clf.fit(X_train, y_train)

    # 예측
    pred = dt_clf.predict(X_test)

    # 정확도
    accuracy.append(accuracy_score(y_test, pred))

print('평균 검증 정확도: ', np.mean(accuracy))
```
