### 리스트
리스트 순차적으로 값을 모아서 저장한다. 리스트를 만들 때는 대괄호([])를 이용한다.

**인덱스는 0부터 시작한다.**

제거할 요소가 여러개일 경우 가장 앞에 있는 요소만 제거된다.

리스트의 길이는  문자형과 마찬가지로 len() 함수를 이용해 얻을 수 있다.

###비교 연산자
두 변수를 비교할때 쓰인다. 해당 조건이 맞으면 참(True) 틀리면 (False)를 반환한다.<br/>

|비교 연산자|문법|의미|
|-----------|----|----|
|==|a == b|a와 b가 같다|
|!=|a != b|a와 b가 다르다|
|>|a > b|a가 b보다 크다(초과)|
|<|a < b|a가 b보다 작다(미만)|
|>=|a >= b|a가 b보다 크거나 같다(이상)|
|<=|a <= b|a가 b보다 작거나 같다(이하)|

```python
a =
b =

print('==: ', a == b)
print('!=: ', a != b)
print('>: ', a > b)
print('<: ', a < b)
print('>=: ', a >= b)
print('<=: ', a <= b)
```

|논 연산자|문법|의미|

|and|a and b|a와 b 모두 True면 True 그렇지 않으면 False|
|or|a or b|a와 b 중 하나라도 True면 True 그렇지 않으면 False |
|not|not a|a가 True면 False, False면 True|
