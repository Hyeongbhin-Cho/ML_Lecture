###결정트리
머신러닝 알고리즘 중 직관적으로 이해하기 쉬운 알고리즘으로 데이터에 있는 규칙을 자동으로 찾에내 트리 기반의 분류 규칙을 만드는 것이다. 스무고개 게임과 유사하다고 볼 수 있다.<br/>
분류 규칙이 많을 수록 분류를 결정하는 방식이 더욱 복잡해 질 것이고 이는 곧 과적합으로 이어진다. 따라서 최대한 균일한 데이터 세트로 분류될 수 있도록 트리를 구성하는 것이 중요하다.

**결정트리 하이퍼파라미터**<br/>
결정 트리를 어떻게 구성할 것지 정할 수 있는 입력 변수이다.
- min_samples_split:
    - 노드를 분할하기 위한 최소한의 샘플 데이터 수로 과적합을 제어하는 데 사용된다.
    - 기본값은 2, 작게 설정할 수록 분할 되는 노드가 많아져서 과적합 가능성이 증가된다.
- min_samples_leaf:
    - 말단 노드가 되기 위한 최소한의 샘플 데이터 수
- max_features:
    - 최적의 분할을 위해 고려할 최대 특징 개수, 기본값은 전체 특징
    - int형이면 대상 특징의 수, float형이면 대상 특징의 비율
- max_depth:
    - 트리의 최대 깊이를 규정한다.
    - 깊이가 깊어지면 과적합 가능성이 증가한다.
- max_leaf_nodes:
    - 말단 노드의 최대 개수

###앙상블 학습(Ensemble Leaning)
앙상블 학습은 여러개의 분류기(Classifier)를 생성하고 각 분류기의 예측을 결합함으로 써보다 정확한 최종 예측을 도출하는 기법이다. 앙상블 학습의 유형은 보팅(Voting), 베깅(Bagging), 부스팅(Boosting) 크게 3가지로 구분할 수 있다.
- 보팅(Voting): 서로 다른 알고리즘으로 된 분류기를 결합해 투표를 진행하여 최종 예측을 도출하는 방법
- 베깅(Bagging): 모두 같은 알고리즘으로 되어있지만 각각 다른 데이터 샘플링을 하여 학습된 분류기를 결합해 투표를 진행하여 최종 예측을 도출하는 방법
- 부스팅(Boosting): 여러 개의 약한 분류기가 순차적으로 학습을 수행하되, 앞에서 학습한 분류기가 예측이 틀린 데이터에 대해서는 올바르게 예측할 수 있도록 다음 분류기에게 가중치를 부여하며 진행되는 방법

####보팅(Voting)
보팅 방법에는 하드보팅(Hard Voting), 소프트 보팅(Soft Voting) 두 가지가 있다.
- 하트 보팅(Hard Voting): 다수결의 원칙으로 다수의 분류기가 예측 결과로 최종 예측 결과를 도출하는 방법
- 소프트 보팅(Soft Voting): 분류기 들의 레이블 값 결정 확률을 모두 더하고 이를 평균해서 가중 높은 확률을 가지는 결과를 최종 예측 결과로 도출하는 방법
