###비교 연산자
두 변수를 비교할때 쓰인다. 해당 조건이 맞으면 참(True) 틀리면 (False)를 반환한다.<br/>

|비교 연산자|문법|의미|
|-----------|----|----|
|==|a == b|a와 b가 같다|
|!=|a != b|a와 b가 다르다|
|>|a > b|a가 b보다 크다(초과)|
|<|a < b|a가 b보다 작다(미만)|
|>=|a >= b|a가 b보다 크거나 같다(이상)|
|<=|a <= b|a가 b보다 작거나 같다(이하)|

```python
a = 5
b = 5

print('==: ', a == b)
print('!=: ', a != b)
print('>: ', a > b)
print('<: ', a < b)
print('>=: ', a >= b)
print('<=: ', a <= b)
```

### 논리 연산자
불린 값의 연산을 할 수 있게끔 한다.<br/>

|논 연산자|문법|의미|
|-----------|----|----|
|and|a and b|a와 b 모두 True면 True 그렇지 않으면 False|
|or|a or b|a와 b 중 하나라도 True면 True 그렇지 않으면 False |
|not|not a|a가 True면 False, False면 True|


```python
a = False
b = False

print('and: ', a and b)
print('or: ', a or b)
print('not: ', not a)
```

###if
if는 조건문으로 만약 조건식이 참(True)이면 아래의 내용을 실행하고 그러지 않으면 넘어간다.

```python
a = 11

if a > 10 :
    print('hello world')
    print("안녕하세요")

print('bye world')
```

if else 구문을 통해 조건식을 만족하지 않을 경우 else 아래의 내용을 실행할 수 있다.

```python
a = 11

if a > 10 :
    print("참외")
else :
    print("수박")

print("사과")
```

if elif else 구문을 통해 여러개의 조건식을 검사할 수 있다.

```python
a = 7

if a > 10 :
    print("참외")
elif a < 6:
    print("복숭아")
else :
    print("수박")
```

### for
for은 반복문으로 주어진 리스트 내의 원소를 하나씩 변수에 넣어 실행한다. <br/>
리스트의 길이 만큼 반복된다.

```python
pocketmon = ['파이리', '이상해씨', '꼬북이']

for partner in pocketmon :
    print(partner)

print("반가워")
```

range() 함수를 이용해 원하는 수의 범위를 만들 수 있다.
- range(끝:<숫자형>): 0부터 (끝-1)까지의 수의 범위를 반환한다.
- range(시작:<숫자형>, 끝:<숫자형>, 간격:<숫자형>):<br/>
시작부터 (끝 - 1)까지의 수의 범위를 반환한다. 간격만큰 숫자 사이의 간격이 정해진다.

```python
for i in range(2, 10, 2):
    print(i)
```

```python
for i in range(1, 10, 2):
    print(i)
```

### while
while 또한 반복문으로 조건식이 참일 경우 계속해서 아래 코드를 반복한다.

```python
a = 0

while a < 10 :
    print(a)
    a = a + 1
```

```python
# [문제]
# '*'로 입력된 수의 높이를 가지는 삼각형을 출력하라
'''
[예시]
input: 5
    *
   ***
  *****
 *******
*********
'''
# 변수 height에 키보드로 입력된 숫자를 저장한다.
height = int(input("숫자를 입력해주세요: "))

# 아래에 코드를 작성하세요
for i in range(height):
    print(" " * (height - 1 - i) + "*" * (1 + 2 * i))
```

### 람다(lambda) 표현식
간단한 함수일 경우 위와 같이 함수를 정의하면 가독성이 떨어진다. 그래서 이를 개선하기 위해 람다(lambda) 표현식이 쓰인다. <br/>

```python
a = [1, 2, 3]
squares = map(lambda x : x**2, a)
list(squares)
```

클래스
------
유사한 형태의 코드를 생성하기 위한 일종의 틀<br/>
ex) 게임의 NPC의 경우 움직임, 대사, 외형이 비슷한 경우가 많다 이들을 효율적으로 생성하기 위해 클래스가 사용된다.<br/>
**클래스는 필드, 생성자, 메소드로 나뉜다.**

```python
# 클래스 선언
class Calculator:
    # 필드
    a = 0
    b = 0

    # 생성자
    def __init__(self, a, b):
        self.a = a
        self.b = b

    # 메소드
    def sum(self):
        return self.a + self.b

cal = Calculator(1, 3)
cal.sum()
```
